```python
#Loading in libraries
import yfinance as yf
import pandas as pd
import matplotlib_inline
```

```python
#Getting TESLA data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1y")  
print(tesla_data)
```

```python
#loading packages for web scrapping
import requests
from bs4 import BeautifulSoup
```

```python
import requests
from bs4 import BeautifulSoup

# Selecting Tesla HTML link to harvest data
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Finding the table
table = soup.find('table')

# Extracting headers
headers = [header.text for header in table.find_all('th')]

# Extracting rows
data = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    data.append([column.text for column in columns])

# Print headers and data
print("Headers:", headers)
for row in data:
    print(row)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(tesla_data.index, tesla_data['Close'], label='Tesla Stock Price')
plt.title('Tesla Stock Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
